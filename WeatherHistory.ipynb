```python
data = data.drop(columns=["Loud_Cover"])
```

## resampling

```python
#upward resampling is used to change hourly data to day data
data = data.resample('D').apply(lambda x: x.mode()[0])
```

```python
data.info()
```

```python
numcols = data.select_dtypes(include=np.number)
objcols = data.select_dtypes(include="object")
```

```python
fig,axes = plt.subplots(nrows=2, ncols=4, figsize=(16,8))
axes = axes.flat

for index, column in enumerate(numcols):
    sns.boxplot(
       data = data,
       x = column,
       ax = axes[index]
    )
```

```python
fig,axes = plt.subplots(nrows=2, ncols=4, figsize=(16,8))
axes = axes.flat

for index, column in enumerate(numcols):
    sns.histplot(
       data = data,
       x = column,
       kde = True,
       ax = axes[index]
    )
```
